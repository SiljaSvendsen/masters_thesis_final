## Functions

```python
def skewed_gaussian(x, N, mu, sigma, alpha):
    Norm = N*binwidth/np.sqrt(2.0*np.pi)/sigma
    z = (x-mu)/sigma
    gaussian = Norm*np.exp(-0.5*z**2)
    cdf_gaussian = norm.cdf(alpha*z)
    return 2*gaussian*cdf_gaussian

def double_skewed_gaussian(x,
                           N1, mu1, sigma1, alpha1,
                           N2, mu2, sigma2, alpha2
                          ):
    return skewed_gaussian(x, N1, mu1, sigma1, alpha1)+skewed_gaussian(x, N2, mu2, sigma2, alpha2)

def akaiki_information_criterion(k, N, residuals):
    """
    Returns Akaiki Information Criterion
    AIC = 2*k - 2*ln(L)
    ln(L) = -0.5*N*(ln(2*pi)+1+ln(RSS))
    RSS = average of sum of squares of the residuals
    
    Input:
    -------
    k: number of fit parameters
    N: number of data points (bins)
    """
    
    max_log_likelihood = -0.5*N*(np.log(2*np.pi) + 1 + np.log(np.sum(residuals**2)/N))
    
    return 2*k - 2*max_log_likelihood


def AIC_small_sample(k, N, residuals):
    AICc = akaiki_information_criterion(k, N, residuals)*((2*k)*(k+1))/(N-k-1)
    return AICc

# Compare AIC Results
def relative_probability_AIC(aic_arr):
    """
    The relative probability that an inferior model is as good as the preferred model.
    """
    aic_min = min(aic_arr)
    zum = np.sum(np.exp(-0.5*(aic_min-aic_arr)))
    
    all_weights = []
    
    for aic in aic_arr:
        
        if aic==aic_min:
            continue
            
            
        x = -0.5*(aic_min-aic)
        if x > 700:  # Threshold for overflow
            exp_result = float('inf')  # Represent overflow
        elif x < -700:  # Threshold for underflow
            exp_result = 0.0  # Approximation for exp(-large_number)
        else:
            exp_result = np.exp(x)
        print(exp_result)
        
        wi = exp_result/zum
        
        all_weights.append(wi)
    return all_weights

def nice_string_text(x,y, minuit_obj, entries, fit_color, fitfunc, aic,font_size_text=12, ax=None):
    '''
    x, y: figure position of string (left lower corner)
    fitfunc: "string"
    aic  = Akeiki Information Criterion
    '''

    #fit information dictionary
    chi2_obj_value = minuit_obj.fval
    N_data_points = len(bin_centers[select])
    Ndof_obj_value = len(bin_centers[select]) - minuit_obj.nfit
    Prob_obj_value = stats.chi2.sf(chi2_obj_value, Ndof_obj_value) # The chi2 probability given N_DOF degrees of freedom
    
    #print(Prob_obj_value)
    
    fit_info = (
        f"Entries: {entries}\n"
        f"Ndatapoint:{N_data_points}\n"
        f"Chi2/ Ndof: {chi2_obj_value:.2f}/ {Ndof_obj_value}\n"
        f"Prob(Chi2, Ndof): {Prob_obj_value:.2f}\n"
        f"AIC: {np.around(aic,1)}"
    )
        
    # fit parameter dictionary
    params_dict = {param: (minuit_obj.values[param], minuit_obj.errors[param])
                   for param in minuit_obj.parameters}

    # Format it into a multi-line string
    fit_params = "\n".join(fr"{key} = {value:.1f} $\pm$ {error:.1f}"
                               for key, (value, error) in params_dict.items())

    # join text
    formatted_text = fitfunc+":\n" + fit_info + "\n" + fit_params


    # Plot and display the text on the figure
    if ax==None:
        fig, ax=plt.subplots(figsize=(5,5))
    else:
        ax=ax
        ax.text(x, y, formatted_text, fontsize=font_size_text, transform=plt.gca().transAxes, color=fit_color)
```

```python
binwidth=1
size=100
x_test = np.linspace(-20,20, size)
y_test = (skewed_gaussian(x_test, N=100, mu=-3, sigma=4, alpha=3)
          + np.random.normal(loc=0, scale=1, size=size))
y_test = (double_skewed_gaussian(x_test, N1=100, mu1=-3.0, sigma1=4.0, alpha1=3,
                                N2=10, mu2=10.0, sigma2=1.0, alpha2=0)
          + np.random.normal(loc=0, scale=1, size=size)
          
         )

y_err = np.std(y_test)
plt.errorbar(x_test, y_test, yerr=y_err, marker=".", label="data")


# fit
chi2_single_peak = Chi2Regression(skewed_gaussian, x_test, y_test, y_err)
minuit_obj_single_peak = Minuit(chi2_single_peak,
                                N=100, mu=-3.0, sigma=4.0, alpha=3
                               )
minuit_obj_single_peak.errordef=1.0 # chi2 fit
minuit_obj_single_peak.migrad()     # run minimisation

#
chi2_double_peak = Chi2Regression(double_skewed_gaussian, x_test, y_test, y_err)
minuit_obj_double_peak = Minuit(chi2_double_peak,
                                N1=100, mu1=-3.0, sigma1=4.0, alpha1=3,
                                N2=10, mu2=10.0, sigma2=1.0, alpha2=0
                               )
minuit_obj_double_peak.errordef=1.0 # chi2 fit
minuit_obj_double_peak.migrad()     # run minimisation

# AIC
residuals_skewed = y_test - skewed_gaussian(x_test, *minuit_obj_single_peak.values[:])
AIC_single=akaiki_information_criterion(k=len(minuit_obj_single_peak.values[:]), N=len(y_test),
                                        residuals=residuals_skewed)

residuals_skewed_double = y_test - double_skewed_gaussian(x_test, *minuit_obj_double_peak.values[:])
AIC_double=akaiki_information_criterion(k=len(minuit_obj_double_peak.values[:]), N=len(y_test),
                                        residuals=residuals_skewed_double)


# plotting fits
plt.plot(x_test, skewed_gaussian(x_test, *minuit_obj_single_peak.values[:]), label="Single Skewed Gaussian")
plt.plot(x_test, double_skewed_gaussian(x_test, *minuit_obj_double_peak.values[:]), label="Double Skewed Gaussian")
print("single peak:", AIC_single)
print("double peak:", AIC_double)
print("weights", relative_probability_AIC([AIC_single, AIC_double]))
plt.legend()
```

# Fit the data
Get fit parameters

```python
# read data
diff_tau_all = pd.read_csv("time_delay_diff_241021.csv")
```

```python
# select data
dataframe=diff_tau_all
network="GNEvGN"
parameter="KaaNG"
date="301124"
scalefactor=4.0
data_color="darkorchid" #"cornflowerblue"#"grey"#"darkorchid"#"darkorange"
font_size=10

mask0 = dataframe["two_networks"]==network
mask1 = dataframe["parameter"]==parameter
mask2 = dataframe["rel change"]==scalefactor

sample = dataframe[mask0 & mask1 & mask2]["time diff"]

Nbins = 2*int(np.sqrt(len(sample))) # 2*33
binwidth = ((12-(-21))/float(Nbins))#(max(sample)-min(sample))/float(Nbins)
print("Nbins",Nbins, "bindwidth", binwidth)


# PREPARE DATA FOR FITTING
binning = np.linspace(-21, 12, Nbins)
counts, bin_edges = np.histogram(sample, bins=binning)


unc_counts = np.sqrt(counts)
bin_centers = bin_edges[:-1]+(bin_edges[1]-bin_edges[0])/2
select = counts > 0


# PLOT DATA
# as aid to set initial guesses
fig, ax = plt.subplots(figsize=(8,6))
ax.errorbar(bin_centers, counts, yerr=unc_counts,
            marker=".", drawstyle="steps-mid",color=data_color,
            label=f"s = {scalefactor}"
           )


# SINGLE SKEWED GAUSSIAN

# chi2 fit
chi2_single_peak = Chi2Regression(skewed_gaussian, bin_centers[select], counts[select], unc_counts[select])
minuit_obj_single_peak = Minuit(chi2_single_peak,
                                N=125, mu=-17, sigma=4.0, alpha=4
                               )
minuit_obj_single_peak.errordef=1.0 # chi2 fit
minuit_obj_single_peak.migrad()     # run minimisation

# AIC
residuals_single_peak = bin_centers[select] - skewed_gaussian(bin_centers[select],
                                                              *minuit_obj_single_peak.values[:])

AIC_single_peak = akaiki_information_criterion(k=len(minuit_obj_single_peak.values[:]),
                                               N=len(bin_centers[select]),
                                               residuals=residuals_single_peak,
                                              )

AICc_single_small = AIC_small_sample(k=len(minuit_obj_single_peak.values[:]),
                                               N=len(bin_centers[select]),
                                               residuals=residuals_single_peak,
                                              )
# DOUBLE SKEWED GAUSSIAN

# chi2 fit
chi2_double_peak = Chi2Regression(double_skewed_gaussian, bin_centers[select], counts[select], unc_counts[select])
minuit_obj_double_peak = Minuit(chi2_double_peak,
                                N1=125, mu1=-17, sigma1=4.0, alpha1=4,
                                N2=25, mu2=3, sigma2=2.0, alpha2=-7.0
                               )

minuit_obj_double_peak.errordef=1.0 # chi2 fit
minuit_obj_double_peak.migrad()     # run minimisation

# AIC
residuals_double_peak = bin_centers[select] - double_skewed_gaussian(bin_centers[select],
                                                              *minuit_obj_double_peak.values[:])

AIC_double_peak = akaiki_information_criterion(k=len(minuit_obj_double_peak.values[:]),
                                               N=len(bin_centers[select]),
                                               residuals=residuals_double_peak,
                                              )
AICc_double_small = AIC_small_sample(k=len(minuit_obj_double_peak.values[:]),
                                               N=len(bin_centers[select]),
                                               residuals=residuals_double_peak,
                                              )


# compare AICs
wi = relative_probability_AIC([AIC_single_peak, AIC_double_peak])


# PLOT FITS
x = np.linspace(-21, 12, 100)
ax.plot(x, skewed_gaussian(x, *minuit_obj_single_peak.values[:]),
       color="red", label="Single Skewed Gaussian")
ax.plot(x, double_skewed_gaussian(x, *minuit_obj_double_peak.values[:]),
       color="black",linestyle= "--", label="Double Skewed Gaussian")

# FIT DATA
nice_string_text(x=0.65, y=0.45, minuit_obj=minuit_obj_single_peak, entries=len(sample), aic=AIC_single_peak,
                 fit_color="red", fitfunc="Single Skewed Gaussian",font_size_text=12, ax=ax)
nice_string_text(x=0.30, y=0.25, minuit_obj=minuit_obj_double_peak, entries=len(sample), aic=AIC_double_peak,
                 fit_color="black", fitfunc="Double Skewed Gaussian",font_size_text=12,ax=ax)


# FIGURE ADJUSTMENTS + SAVE
ax.set_title(f"{network}_{parameter}_{scalefactor}\n Relative Prob(AIC)={wi}", fontsize=font_size)
ax.set_xlabel(r"$\Delta\tau_{GNE-GN}$", fontsize=font_size)
ax.set_ylabel(f"Frequency/{np.around(binwidth,1)}", fontsize=font_size)
plt.legend(fontsize=12, loc="upper right")

ax.set_xlim(-22, 13)
ax.set_ylim(0,200)
plt.tight_layout()

print(f"{network}_{parameter}_{scalefactor}")
print(f"AIC bias corrected, date: {date}")
print("AICc single peak:", AICc_single_small)
print("AICc double peak", AICc_double_small)
aics = np.array([AICc_single_small, AICc_double_small])

weight = relative_probability_AIC(aics)
#plt.savefig(f"{network}_{parameter}_{scalefactor}_{date}.pdf", dpi=600)
```
