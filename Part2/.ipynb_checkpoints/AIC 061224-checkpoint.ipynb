```python
import numpy as np               # sum, log, pi
import matplotlib.pyplot as plt  # plotting
from scipy import stats          # compute p-values from chi2 and Ndof
from scipy.stats import norm     # defining fit functions
from scipy.integrate import quad # compute pdf
import pandas as pd              # read data
from iminuit import Minuit       # fitting
from Module_241021 import (load_data_w_glob)
import os, sys 
sys.path.append('/Users/siljaxianpallisgaardsvendsen/Documents/KU_K/8_AppStat/AppStat2023')
from ExternalFunctionsNew import Chi2Regression # chi2 minimisation
```

```python
def AIC_bias_corrected(Nparams, Nsample, residuals):
    """
    Returns the bias corrected AIC
    Input: number of model parameters, sample size, and residuals
    """
    def AIC(Nparams, Nsample, residuals):
        maxloglikelihood=-0.5*Nsample*(np.log(2*np.pi)+1+np.log(np.sum(residuals**2)/Nsample))
        AIC=2*Nparams-2*maxloglikelihood
        return AIC
    AICc = AIC(Nparams, Nsample, residuals)*((2*Nparams)*(Nparams+1))/(Nsample-Nparams-1)
    return AICc

def weight(aic):
    """compares only two models, else i should have used a for loop to write the sum"""
    print(min(aic), max(aic), aic)
    return 1/(1 + np.exp(-0.5*(min(aic)-max(aic))))

def gaussian(x, mu, sigma):
    "gaussian pdf"
    return (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-((x-mu)**2)/(2*sigma**2))

def double_gaussian(x, w1, mu1, sigma1, mu2, sigma2):
    "guassian pdf + gaussian pdf"
    return w1*gaussian(x, mu1, sigma1) + (1-w1)*gaussian(x, mu2, sigma2)
```

```python
# read data
diff_tau_all = pd.read_csv("time_delay_diff_241021.csv")
```

```python
def fits_aicc_plot(dataframe = diff_tau_all,
                   network="GNEvGN",
                   parameter="KaaNG",
                   date="061224",
                   scalefactor=1.0,
                   data_color="grey", #----settings for plotting
                   label_fz=18,
                   title_axis_fz=30,
                   minx=-21,
                   maxx= 12, # ensure equal bin widths
                   mu    = 0, #--- initial guesses
                   sigma = 2,
                   w1    = 0.5,
                   mu1   = 0,
                   sigma1= 1,
                   mu2   = 1,
                   sigma2= 1,
                   savefig=False
                  ):
    
    """
    The function plots distributions, fits distributions, and calculates the bias-corrected AIC
    It is a very case specific function.
    Fx. ensure that the dataframe has categories "two_networks", "parameter", "rel change"
    minx = min x-value among all samples
    maxx = max x-value among all samples
    """
    

    #---- select the data
    mask0 = dataframe["two_networks"]==network
    mask1 = dataframe["parameter"]==parameter
    mask2 = dataframe["rel change"]==scalefactor
    sample = dataframe[mask0 & mask1 & mask2]["time diff"]

    # settings for fitting
    Nbins = 2*int(np.sqrt(len(sample))) #2*33
    binwidth=(maxx-minx)/float(Nbins)
    
        
    binning = np.linspace(minx, maxx, Nbins)
    counts, bin_edges = np.histogram(sample, bins=binning)
    unc_counts = np.sqrt(counts)
    counts_norm = counts/(len(sample)*binwidth)
    unc_counts_norm = unc_counts/(len(sample)*binwidth)

    bin_centers= bin_edges[:-1]+(bin_edges[1]-bin_edges[0])/2
    
    select=counts>0

    # plot the sample
    fig, ax = plt.subplots(figsize=(8,6))
    ax.errorbar(bin_centers[select], counts_norm[select], yerr=unc_counts_norm[select],
                marker=".", drawstyle="steps-mid", color=data_color, label=r"$N = $"+f"{len(counts_norm[select])}"
               )

    # --- perform the fits + calculate AICc

    # SINGLE PEAK
    chi2_guass = Chi2Regression(gaussian,
                                bin_centers[select],
                                counts_norm[select],
                                unc_counts_norm[select])
    minuit_obj_gauss = Minuit(chi2_guass,
                              mu=mu,
                              sigma=sigma
                             )
    minuit_obj_gauss.errordef=1.0 #chi2
    minuit_obj_gauss.migrad()     # run minimisation

    # bias-corrected AIC
    residual_gauss = counts_norm[select]-gaussian(bin_centers[select],*minuit_obj_gauss.values[:])
    AICc_gauss = AIC_bias_corrected(Nparams=len(minuit_obj_gauss.values[:]),
                                    Nsample=len(counts_norm[select]),
                                    residuals=residual_gauss
                                   )


    # DOUBLE PEAK
    chi2_doublegauss = Chi2Regression(double_gaussian,
                                bin_centers[select],
                                counts_norm[select],
                                unc_counts_norm[select])
    minuit_obj_doublegauss = Minuit(chi2_doublegauss,
                                    w1=w1,
                                    mu1=mu1,
                                    sigma1=sigma1,
                                    mu2=mu2,
                                    sigma2=sigma2
                                   )
    minuit_obj_doublegauss.errordef=1.0 #chi2
    minuit_obj_doublegauss.migrad()     # run minimisation

    # bias-corrected AIC
    residual_doublegauss = counts_norm[select]-double_gaussian(bin_centers[select],*minuit_obj_doublegauss.values[:])
    AICc_doublegauss = AIC_bias_corrected(Nparams=len(minuit_obj_doublegauss.values[:]),
                                    Nsample=len(counts_norm[select]),
                                    residuals=residual_doublegauss
                                   )


    # ---- plot fits
    x = np.linspace(minx, maxx, 500)
    y1 = gaussian(x, *minuit_obj_gauss.values[:])
    y2 = double_gaussian(x, *minuit_obj_doublegauss.values[:])

    ax.plot(x, y1,
            color="red",
            label=("Gaussian:\n $\mu$ = "
                   +f"{minuit_obj_gauss.values['mu']:.1f}"
                   +"$\pm$"
                   +f"{minuit_obj_gauss.errors['mu']:.1f}"
                   +", $\sigma$ = "
                   +f"{minuit_obj_gauss.values['sigma']:.1f}"
                   +"$\pm$"
                   +f"{minuit_obj_gauss.errors['sigma']:.1f}"
                  )
           )

    ax.plot(x, y2,
            color="black",
            label="Double Gaussian"
           )

    # ---- plot, adjustments
    ax.set_title(f"Scale Factor = {scalefactor}", fontsize=title_axis_fz)
    ax.set_xlabel(r"$\Delta\tau_{GNE-GN}$", fontsize=title_axis_fz)
    ax.set_ylabel("Relative Frequency", fontsize=title_axis_fz)
    ax.tick_params(axis='both', which='major', labelsize=label_fz)
    ax.set_xlim(minx-0.5, maxx+0.5)
    ax.set_ylim(0,0.5)
    plt.legend(fontsize=label_fz)
    plt.tight_layout()
    if savefig:
        plt.savefig(f"distribution_{network}_{parameter}_{scalefactor}_{date}.pdf", dpi=600)
        
    
        
    # ---
    chi2_obj_value_g = minuit_obj_gauss.fval
    Ndof_obj_value_g = len(bin_centers[select]) - minuit_obj_gauss.nfit
    Prob_obj_value_g = stats.chi2.sf(chi2_obj_value_g, Ndof_obj_value_g)
    
    chi2_obj_value_gg = minuit_obj_doublegauss.fval
    Ndof_obj_value_gg = len(bin_centers[select]) - minuit_obj_doublegauss.nfit
    Prob_obj_value_gg = stats.chi2.sf(chi2_obj_value_gg, Ndof_obj_value_gg)
    
    
    fit_info = f"filename: distribution_{network}_{parameter}_{scalefactor}_{date}.txt"
    with open(fit_info, "w") as f:
        sys.stdout = f
        
        print(f"Number of bins = {Nbins}. Bin width = {binwidth}")
        print((f"\nAICc Gauss= {AICc_gauss:.2f}"+
           "\n"+
           f"chi2/Ndof = {chi2_obj_value_g}/{Ndof_obj_value_g}"+
           "\n"+
           f"prob(chi2, Ndof)={Prob_obj_value_g}"+
           "\n"+
           "mu = "+f"{minuit_obj_gauss.values['mu']:.2f} \pm {minuit_obj_gauss.errors['mu']:.2f}"+
                   "\n"+
                   r"\sigma = "+f"{minuit_obj_gauss.values['sigma']:.2f} \pm {minuit_obj_gauss.errors['sigma']:.2f}"
                  ))
        print(f"\nAICc Double Gauss = {AICc_doublegauss:.2f}"
                   +"\n"+
           f"chi2/Ndof = {chi2_obj_value_gg}/{Ndof_obj_value_gg}"+
           "\n"+
           f"prob(chi2, Ndof)={Prob_obj_value_gg}"+
          "\n"+
          "w_1 = "+f"{minuit_obj_doublegauss.values['w1']:.2f} \pm {minuit_obj_doublegauss.errors['w1']:.2f}"
                   +"\n"+
                   "\mu_1 = "+f"{minuit_obj_doublegauss.values['mu1']:.2f} \pm {minuit_obj_doublegauss.errors['mu1']:.2f}"
                   +"\n"+
                   r"\sigma_1 = "+f"{minuit_obj_doublegauss.values['sigma1']:.2f} \pm {minuit_obj_doublegauss.errors['sigma1']:.2f}"
                   +"\n"+
                   r"\mu_2 = "+f"{minuit_obj_doublegauss.values['mu2']:.2f} \pm {minuit_obj_doublegauss.errors['mu2']:.2f}"
                   +"\n"+
                   r"\sigma_2 = "+f"{minuit_obj_doublegauss.values['sigma2']:.2f} \pm {minuit_obj_doublegauss.errors['sigma2']:.2f}"
                  )
        print(f"\nweight_factor = {weight([AICc_gauss, AICc_doublegauss])}")
        sys.stdout = sys.__stdout__
        
    return AICc_gauss, AICc_doublegauss
```

```python
fits_aicc_plot(savefig=True)
```
