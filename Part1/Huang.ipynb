### 1. Investigate the kaa-to-kmi ratio
In a previous notebook: "Huang et al initial investigation", I plotted phase spaces for $\frac{k_{aa}}{k_{mi}}=0.25, 0.5, 1.0, 2.0$. I learned that when the auto-activation is weak (high $k_{aa}$), the tri-stability breaks down. To identify at which ratio, tristabilitiy breaks down, I will make a bifurcation diagram. One where I use the Huang model (OR-logic), and one with the modified version (AND-logic)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1,2,figsize=(8,6))
tristability=func_bifurcation_diagram_one_par(model_2D=equations_huang_or_logic, par=p0_huang,
                                              key="kaa1", key2="kaa2",
                                              filename="Bifurcation diagram Huang OR logic",
                                              par_bifur_start=0.125, par_bifur_stop=1, par_bifur_stepsize=0.025,
                                              radius=0.4, curve_step=300,
                                              par_bifur_log2=False,
                                              commontitle="Huang OR-logic",
                                              right_ylabel="[GATA6]", left_ylabel="[NANOG]",
                                              font_size=30,
                                              save_fig=True,
                                              return_tristable=True,
                                              fig_global=fig,
                                              axs_global=ax,
                                              x_ticks_pos=np.around(np.linspace(0.125, 1, 10),1),
                                              x_ticks=np.around(np.linspace(0.125, 1, 10)/0.5,1),
                                              x_start=-0.1, y_start=-0.1, x_stop=2.5, y_stop=2.5,
                                              xlabel=r"$\frac{k_{aa1}}{k_{mi1}}=\frac{k_{aa2}}{k_{mi2}}=\frac{k_{aa}}{k_{mi}}$"
                                             )
```

```python
print("The Huang model is tristable when kaa/kmi=",2*np.array(tristability))
```
