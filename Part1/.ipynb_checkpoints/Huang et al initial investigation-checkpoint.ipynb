**Conclusion**

Tristability does not depend only on the logical architecture. <br>

The auto-activation loop has to be strong enough.<br> For example,
When kma/kmi = 1/0.5, and (AA + MI), the system becomes bi-stable in the Epi and PrE state. <br>
When kma/kmi = 1/0.5, and (AA * MI), the system becomes mono-stable in DN. <br>


When kma/kmi = 0.5/0.5, tri-stability is possible. The logical architecture determines the stability of the states.
If protein x1, x2 is regulated by AA and MI (AA * MI), the system is stable in Epi, PrE, and DN. <br>
If Protein x1, x2 is regulated by AA or MI (AA + MI), the system is stable in Epi, PrE, and DP. <br>
